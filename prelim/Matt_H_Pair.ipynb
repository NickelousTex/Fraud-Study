## Modeling

```python
clean = pd.read_csv('data/cleaned_data.csv')
```

```python
clean.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
clean
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = clean.iloc[:,1:]
y = clean.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.3)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import recall_score, make_scorer
from sklearn.grid_search import GridSearchCV
```

```python
recaller = make_scorer(recall_score)
```

```python
grad = GradientBoostingClassifier()

gb_grid_params = {
              'n_estimators': [60,100,150],
              'learning_rate': [0.1, 0.05, 0.02, 0.01],
              'max_depth': [4, 6, 8],
              'min_samples_leaf': [20, 50,100,150],
              'max_features': [1.0, 0.3, 0.1] 
              }
gridsearch = GridSearchCV(GradientBoostingClassifier(), gb_grid_params, scoring=recaller, n_jobs=-1,verbose=3)
```

```python
gb_grid_params2 = {
              'n_estimators': [60,100,150,300],
              'learning_rate': [0.1, 0.05, .15, .2],
              'max_depth': [4,8,10,12],
              'min_samples_leaf': [15,20, 50,100,150],
              'max_features': [1.0, 0.3,.5, 0.1] 
              }
gridsearch2 = GridSearchCV(GradientBoostingClassifier(), gb_grid_params2, scoring=recaller, n_jobs=-1,verbose=3)
```

```python
##Careful Running!
# gridsearch.fit(X_train,y_train)
```

```python
## Careful Running!
# gridsearch2.fit(X_train,y_train)
```

```python
y_test_preds = grad.predict(X_test)
y_train_preds = grad.predict(X_train)
```

```python
grad.score(X_test, y_test)
```

```python
recall_score(y_test, y_test_preds)
```

```python
gridsearch.best_score_
```

```python
gridsearch.best_params_
```

```python
best_grad = gridsearch.best_estimator_
```

```python
best_grad.fit(X_train,y_train)
```

```python
recaller(best_grad, X_test, y_test)
```

```python
best_test_pred = best_grad.predict(X_test)
```

```python
recall_score(y_test, best_test_pred)
```

## Gridsearched model

```python
best_grad2 = gridsearch2.best_estimator_
best_grad2.fit(X_train, y_train)
```

```python
recaller(best_grad2, X_test, y_test)
```
